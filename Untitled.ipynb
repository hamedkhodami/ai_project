```python
import pandas as pd
import numpy as np
import random
```

```python
num_records= 100
```

```python
location= ['منطقه یک','منطقه دو','منطقه سه','منطقه چهار','منطقه پنج']
Area=[random.randint(50,250) for _item in range(num_records)]
Rooms=[random.randint(1,5) for _item in range(num_records)]
Floors=[random.randint(1,12) for _item in range(num_records)]
Location_choices=[random.choice(location) for _item in range(num_records)]
```

```python
bas = {
    'منطقه یک':1000000000,
    'منطقه دو':5000000000,
    'منطقه سه':2000000000,
    'منطقه چهار':3000000000,
    'منطقه پنج':4000000000
}
```

```python
def calculate_price(area,rooms,floor,location):
    base_price =bas[location]
    area_factor = (area / 50)*100000000
    room_factor=rooms*200000000
    floor_factor=(12-floor)*50000000
    return base_price +area_factor +room_factor +floor_factor
```

```python
Price=[calculate_price(Area[i],Rooms[i],Floors[i],Location_choices[i])for i in range(num_records)]
formated_price=["{:,}".format(int(price)) for price in Price]
```

```python
data={'Area':Area,'Rooms':Rooms,'Floors':Floors,'location':Location_choices,'Price':formated_price}
```
